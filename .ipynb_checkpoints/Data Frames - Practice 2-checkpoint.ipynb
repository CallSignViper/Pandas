### Changing Data Types

```python
import pandas as pd
```

```python
pd.read_csv("employees.csv").info()
```

```python
df = pd.read_csv('employees.csv',parse_dates = ['Start Date', 'Last Login Time'])
df.head(3)
```

```python
df['Start Date'].head(3)
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date']).head(3)
```

```python
df['Last Login Time'] = pd.to_datetime(df['Last Login Time'])
```

```python
df['Senior Management'] = df['Senior Management'].astype('bool')
```

```python
df['Gender'] = df['Gender'].astype('category')
```

### Filter A DataFrame Based On A Condition 

```python
df[df['Gender'] == 'Male']
```

```python
df[df['Team']=='Finance']
```

```python
finance = df['Team'] == 'Finance'
df[finance]
```

```python
df['Senior Management'] == True
```

```python
marketing = df['Team'] != 'Marketing'
df[marketing]
```

```python
df[df['Salary'] > 110000]
```

```python
bonus=df['Bonus %'] < 1.5
df[bonus]
```

```python
startdate85=df['Start Date'] <= '1985-01-01'
df[startdate85]
```

### Filter with More than Once Condition (AND)

```python
male = df["Gender"] == 'Male'
marketing = df['Team']=='Marketing'
```

```python
df[male & marketing]
```

### Filter with More than One Condition (OR)

```python
df[male | marketing]
```

```python
Senior_Management = df['Senior Management']
SD = df['Start Date'] < "1990-01-01"

df[Senior_Management | SD]
```

```python
Robert = df['First Name'] == 'Robert'

Client_Services = df['Team'] == 'Client Services'

SD2 = df['Start Date'] > "2016-06-01"

df[(Robert & Client_Services) | SD2]
```

###  The .isin() Method

### The .isnull() and .notnull() Methods

### The .between() Method

```python
salary_analysis = df['Salary'].between(60000,70000)
df[salary_analysis].head()
```
