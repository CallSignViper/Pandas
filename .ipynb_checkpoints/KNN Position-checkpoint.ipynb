```python
import pandas as pd
```

```python
df = pd.read_excel('Position Classifier.xlsm')
```

```python
df.head()
```

```python
df.drop(labels='Rk',axis=1,inplace=True)
```

```python
df.head(30)
```

```python
def pos_cat(position):
    return position.upper()

def pos_len(position):
    
    if len(position)>2:
        position2 = position.split("/")
        return position2[1]
    else:
        return position
        
```

```python
df['Pos'] = df['Pos'].apply(pos_cat)
df['Pos'] = df['Pos'].apply(pos_len)
```

```python
df.head(30)
```

```python
df.drop('Pos1',axis=1,inplace=True)
```

```python
df.head(30)
```

```python
writer = pd.ExcelWriter(path="U:\Pandas\'Clean Data'",engine='xlsxwriter')
```

```python
df.to_excel(writer,sheet_name='Clean Data')
writer.save()
```
