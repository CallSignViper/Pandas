```python
import pandas as pd
import datetime as dt
from pandas_datareader import data
from pandas.tseries.offsets import *
```

```python
company = 'ETH'
start = '2018-01-01'
end = '2018-08-13'

stocks = data.DataReader(name = company, data_source='robinhood',start = start, end=end)
stocks.head()
```

```python
stocks.reset_index('begins_at',inplace=True)
```

```python
stocks.drop('symbol',inplace=True)
```

```python
stocks.set_index('begins_at',inplace=True)
```

```python
stocks.loc['2017-08-14']
```

```python
stocks.iloc[0]
```

```python
daterange = pd.date_range(start='2017-08-15',end='2017-09-15',freq= 'D')
```

```python
mask=stocks.index.isin(daterange)
```

```python
stocks[mask]
```
