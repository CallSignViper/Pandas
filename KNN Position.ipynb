```python
def pos_cat(position):
    return position.upper()

def pos_len(position):
    
    if len(position)>2:
        position2 = position.split("/")
        return position2[1]
    else:
        return position
        
```

```python
pc = pd.read_excel('Position Classifier.xlsm',sheetname='Source Data')
```

```python
pc.drop(labels='Rk',axis=1,inplace=True)
```

```python
pc['Pos'] = pc['Pos'].apply(pos_cat)
pc['Pos'] = pc['Pos'].apply(pos_len)
```
