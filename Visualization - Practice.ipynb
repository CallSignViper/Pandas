### The .plot() Method

```python
nba = pd.read_csv('nba.csv')
```

```python
thunder = nba['Team'] == 'Oklahoma City Thunder'
```

```python
thunderPlot =nba[thunder]
```

```python
thunderPlot
```

```python
thunderPlot['Salary'].plot()
```

```python
thunderPlot.sort_values('Salary',ascending=False)
```
